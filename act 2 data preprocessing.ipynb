```python
df_q1=df.loc[2:11]
df_q1=df_q1.reset_index(drop=True)
df_q1
```

```python

# Convert 'created-at' to datetime format
df_q1['created_at'] = pd.to_datetime(df_q1['created_at'])

# Masking created-at to show Day of the Week, Month, and Year
df_q1['created_at4'] = df_q1['created_at'].apply(lambda x: x.strftime('%A, %B %d'))

# Create 'newtweet-id' by extracting the last 5 digits and adding the reverse of the rest
df_q1['new_tweet_id'] = df_q1['tweet_id'].apply(lambda x: str(x)[-5:] + str(x)[:-5][::-1])

df_q1['username']='yyy'

df_q1[['new_tweet_id','created_at4','username']]
```
